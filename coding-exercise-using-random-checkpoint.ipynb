```python
import random
from typing import List, Tuple

def generate_tickets(ticket_count: int, max_number: int) -> Tuple[List[int], int]:
    """
    Generates a list of unique random lottery tickets and selects one winning ticket.

    Args:
        ticket_count (int): The number of unique tickets to generate.
        max_number (int): The exclusive upper bound for the random numbers
                          (numbers will be in the range [0, max_number)).

    Returns:
        Tuple[List[int], int]: A tuple containing:
            1. The list of generated unique tickets (List[int]).
            2. The single winning ticket selected from that list (int).

    Raises:
        ValueError: If ticket_count is greater than max_number, as it's impossible
                    to sample that many unique numbers from the range.
    """
    # 1. Generate the list of unique tickets using random.sample.
    # We sample 'ticket_count' unique elements from the population of numbers
    # provided by range(max_number).
    try:
        tickets = random.sample(range(max_number), ticket_count)
    except ValueError as e:
        print(f"Error: Cannot generate {ticket_count} unique tickets from a range of {max_number} numbers.")
        raise e

    # 2. Choose one random element from the generated list of tickets.
    winning_ticket = random.choice(tickets)

    # 3. Return the generated list and the winning number as a tuple.
    return (tickets, winning_ticket)

# --- Example Usage ---

# Example 1: Generate 5 unique tickets from numbers 0 up to (but not including) 10.
TICKET_COUNT_1 = 5
MAX_NUMBER_1 = 10
print(f"--- Lottery Draw Setup 1: {TICKET_COUNT_1} tickets from 0-{MAX_NUMBER_1-1} ---")
tickets_1, winner_1 = generate_tickets(TICKET_COUNT_1, MAX_NUMBER_1)

print(f"Generated Unique Tickets: {tickets_1}")
print(f"The Winning Ticket is: {winner_1}")
print("-" * 40)

# Example 2: A slightly larger draw.
TICKET_COUNT_2 = 10
MAX_NUMBER_2 = 50
print(f"--- Lottery Draw Setup 2: {TICKET_COUNT_2} tickets from 0-{MAX_NUMBER_2-1} ---")
tickets_2, winner_2 = generate_tickets(TICKET_COUNT_2, MAX_NUMBER_2)

print(f"Generated Unique Tickets: {tickets_2}")
print(f"The Winning Ticket is: {winner_2}")
print("-" * 40)
```
