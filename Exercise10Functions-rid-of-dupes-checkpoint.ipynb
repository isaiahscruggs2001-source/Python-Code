```python
def unique(input_list=[]):
    """
    Accepts a list and returns a new list containing only the unique elements 
    from the input list, preserving the original order of the first occurrence 
    of each element.

    :param input_list: The list to process (defaults to an empty list).
    :return: A new list with duplicates removed, maintaining order.
    """
    # Use a dictionary or a set to keep track of elements already seen for O(1) average time complexity.
    # We use a set here for simple membership checking.
    seen = set()
    result = []
    
    for element in input_list:
        # Check if the element has already been encountered
        if element not in seen:
            # If it's a new element, add it to the result list
            result.append(element)
            # Mark it as seen
            seen.add(element)
            
    return result

# --- Examples of Usage ---

# Example 1: Numbers
list_of_numbers = [1, 1, 4, 5, 1, 9, 4, 10, 5, 1]
unique_numbers = unique(list_of_numbers)
print(f"Original list (numbers): {list_of_numbers}")
# Expected output: [1, 4, 5, 9, 10]
print(f"Unique elements (numbers): {unique_numbers}") 

# Example 2: Strings
list_of_names = ['Mark', 'Mark', 'John', 'Anne', 'Mark', 'John', 'Susan']
unique_names = unique(list_of_names)
print(f"\nOriginal list (names): {list_of_names}")
# Expected output: ['Mark', 'John', 'Anne', 'Susan']
print(f"Unique elements (names): {unique_names}")

# Example 3: Default empty list
empty_list_unique = unique()
print(f"\nUnique elements (default empty list): {empty_list_unique}")
```
