```python
class Dog:
 def __init__(self, breed='none'):
  self.breed = breed
 
 def set(self, breed='labrador'):
  self.breed = breed
  return self.breed
 
puppy = Dog()
friend = puppy
friend.set()
print(puppy.breed)
```
