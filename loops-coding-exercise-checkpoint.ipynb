```python
secret_year = 1994
print('''
===========================
=== PYTHON GUESSING GAME ===
===========================
''')

# Initial input outside the loop
user_input = int(input('When was Python 1.0 released?: '))

# Loop continues as long as the input is NOT equal to the secret_year
while user_input != secret_year:
    if user_input > secret_year:
        print('It was earlier than that! Try again.')
    elif user_input < secret_year:
        print('It was later than that! Try again.')
    
    # After giving a hint, ask for new input to continue the loop
    user_input = int(input('When was Python 1.0 released?: '))

# This line is only reached when the while loop condition (user_input != secret_year) is false
print('Correct! Python 1.0 was released in 1994.')
```
