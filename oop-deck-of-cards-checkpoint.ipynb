```python
import random

class Card:
    """
    Represents a single playing card with a suit and a value.
    """
    def __init__(self, suit, value):
        """Initializes a Card object."""
        self.suit = suit
        self.value = value

    def present(self):
        """Returns a string representation of the card."""
        return f"{self.value} of {self.suit}"

class Deck:
    """
    Represents a standard 52-card deck.
    """
    def __init__(self):
        """Initializes a full deck of 52 cards."""
        suits = ['hearts', 'diamonds', 'clubs', 'spades']
        values = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']
        # Use a list comprehension to create a full deck of 52 card objects
        self.cards = [Card(suit, value) for suit in suits for value in values]

    def shuffle(self):
        """Randomly shuffles the cards in the deck."""
        random.shuffle(self.cards)

    def deal(self):
        """
        Deals one card from the top of the deck.
        Returns the card object or None if the deck is empty.
        """
        if self.cards:
            # The pop() method removes and returns the last item in a list
            return self.cards.pop()
        else:
            return None

    def count_remaining(self):
        """Returns the number of cards left in the deck."""
        return len(self.cards)

    def get_remaining(self):
        """Returns a list of strings representing the remaining cards."""
        return [card.present() for card in self.cards]

# --- Example Usage ---
if __name__ == "__main__":
    # 1. Create a new deck
    my_deck = Deck()
    print(f"Created a new deck with {my_deck.count_remaining()} cards.")
    
    # 2. Shuffle the deck
    my_deck.shuffle()
    print("The deck has been shuffled.")

    # 3. Deal 5 cards
    print("\nDealing 5 cards:")
    hand = []
    for _ in range(5):
        dealt_card = my_deck.deal()
        if dealt_card:
            # Use the present() method to display the card
            print(f"Dealt: {dealt_card.present()}")
            hand.append(dealt_card)
        else:
            print("No cards left in the deck!")
            break
            
    # 4. Check how many cards are left
    print(f"\nThere are now {my_deck.count_remaining()} cards left in the deck.")

    # 5. Show the first 10 remaining cards
    remaining = my_deck.get_remaining()
    print("\nFirst 10 cards remaining in the deck:")
    print(remaining[:10])
```
