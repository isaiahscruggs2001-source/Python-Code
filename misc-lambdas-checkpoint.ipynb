```python
def apply_func(elements, func):
    for element in elements:
        print(func(element))
```

```python
my_func = lambda x: x * x
apply_func([1,2,3,4,5], my_func)
```

```python
my_func = lambda x: 1 / x
apply_func([1,2,3,4,5], my_func)
```

```python
my_func = lambda x: 0
apply_func([1,2,3,4,5], my_func)
```

```python
apply_func([1,2,3,4,5], lambda x: x * x)
```
