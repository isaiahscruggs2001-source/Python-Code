```python
import requests
import json

# Stack Overflow API endpoint for recent questions
url = 'https://api.stackexchange.com/2.3/questions'
params = {
    'order': 'desc',
    'sort': 'activity',
    'site': 'stackoverflow',
    'pagesize': 5  # Get 5 recent questions
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    # Stack Overflow API returns data in an 'items' array
    questions = data['items']
    
    for question in questions:
        print(f"Title: {question['title']}")
        print(f"Tags: {', '.join(question['tags'])}")
        print(f"Score: {question['score']}")
        print(f"Link: {question['link']}")
        print("-" * 50)
else:
    print(f"Error: {response.status_code}")
```
