```python
numbers = []
for i in range(1, 101):
    numbers.append(i)
```

```python
print('Numbers:', numbers)
```

```python
numbers = [i for i in range(1, 101)]
```

```python
print('Numbers:', numbers)
```

```python
numbers = [i for i in range(1, 101) if i % 3 != 0]
```

```python
print('Numbers:', numbers)
```

```python
numbers = [0 if i % 2 == 0 else 1 for i in range(100)]
```

```python
print('Numbers:', numbers)
```
