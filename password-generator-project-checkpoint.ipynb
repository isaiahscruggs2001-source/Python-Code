```python
import random

def generate_password():
    """
    Handles the password generation procedure based on user-defined settings.
    """
    # Define the character groups that can be used
    letters = 'abcdefghijklmnopqrstuvwxyz' #
    digits = '0123456789' #
    special_chars = '!@#$%&*^|()_+' #

    # --- Get User Settings ---
    try:
        # Ask the user for the desired password length
        length = int(input("Provide password length: "))
        if length <= 0:
            print("Error: Password length must be a positive number.")
            return
    except ValueError:
        print("Error: Please enter a valid number for the length.")
        return

    # Ask the user which character sets to include
    use_upper = input("Use uppercase letters? (y/n): ").lower() == 'y' #
    use_digits = input("Use digits? (y/n): ").lower() == 'y' #
    use_specials = input("Use special characters? (y/n): ").lower() == 'y' #

    # --- Build the Character Pool ---
    # Start with lowercase letters, which are always included
    char_pool = letters

    if use_upper:
        char_pool += letters.upper()
    if use_digits:
        char_pool += digits
    if use_specials:
        char_pool += special_chars

    # --- Generate and Display the Password ---
    # Create the password by picking random characters from the pool
    password = ""
    for _ in range(length):
        # The random.choice() function picks a random character
        password += random.choice(char_pool)

    print(f"Generated password: {password}") #

def main():
    """
    Displays the main menu and handles user navigation.
    """
    while True:
        # Display the main menu options to the user
        print("\nPassword generator") #
        print("Choose option:") #
        print("1: generate password") #
        print("2: exit the program") #
        
        # Accept input from the user
        choice = input("Your choice: ") #

        if choice == '1':
            # Begin the password generation procedure
            generate_password()
            # After completion, the main menu is shown again
        elif choice == '2':
            # Show a farewell message and finish the program
            print("Bye!") #
            break
        else:
            # Show a warning for incorrect input
            print("Please enter a correct value") #

# Start the program by calling the main function
if __name__ == "__main__":
    main()
```
