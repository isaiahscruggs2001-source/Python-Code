```python
import math

for name in dir(math):
    print(name, end='\t')
```

```python
# ceil, floor(), trunc()
print(math.ceil(3.6))
print(math.floor(3.6))
print(math.trunc(3.6))
```

```python
# ceil() always rounds numbers upward to the nearest integer, never less than the number itself
print(math.ceil(3.1))
print(math.ceil(3.505))
print(math.ceil(3.945))
print(math.ceil(3.0))
print(math.ceil(-5.4))
```

```python
# floor() always rounds the number downward to the nearest integer, never greater than the number itself
print(math.floor(3.1))
print(math.floor(3.505))
print(math.floor(3.945))
print(math.floor(3.0))
print(math.floor(-5.4))
```

```python
# trunc() deletes the numbers that come after the decimal and returns the integer
print(math.trunc(3.6))
print(math.trunc(3.505))
print(math.trunc(3.945))
print(math.trunc(3.0))
print(math.trunc(-5.4))
```

```python
# factorial() multiplication of all positive integers less than or equal to the given number
print(math.factorial(3))
print(math.factorial(4))
```

```python
# sqrt() returns the square root of a given number.
print(math.sqrt(16))
print(math.sqrt(100))
```

```python
# hypot() geometry: searching for the hypotenuse in triangles (rare usage)
print(math.hypot(3,4))
```
